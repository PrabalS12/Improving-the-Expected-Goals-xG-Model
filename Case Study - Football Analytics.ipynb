```python
first_sub = df[(df['type_name']=='Substitution') & (df['team_name']=='Manchester United')].index[0]
```

```python
yanited_passingNetwork = df[(df.type_name=='Pass') & 
   (df['team_name']=='Manchester United') & (df.index < first_sub)][['x','y','end_x','end_y','player_name','pass_recipient_name']]
```

```python
#yanited_passingNetwork['player_name'].apply(lambda x: x.split()[0])
```

```python
passingNetwork_df = pd.DataFrame()
for i, name in enumerate(yanited_passingNetwork["player_name"].unique()):
    passx = yanited_passingNetwork.loc[yanited_passingNetwork["player_name"] == name]["x"].to_numpy()
    recx = yanited_passingNetwork.loc[yanited_passingNetwork["pass_recipient_name"] == name]["end_x"].to_numpy()
    passy = yanited_passingNetwork.loc[yanited_passingNetwork["player_name"] == name]["y"].to_numpy()
    recy = yanited_passingNetwork.loc[yanited_passingNetwork["pass_recipient_name"] == name]["end_y"].to_numpy()
    passingNetwork_df.at[i, "player_name"] = name
    #make sure that x and y location for each circle representing the player is the average of passes and receptions
    passingNetwork_df.at[i, "x"] = np.mean(np.concatenate([passx, recx]))
    passingNetwork_df.at[i, "y"] = np.mean(np.concatenate([passy, recy]))
    #calculate number of passes
    passingNetwork_df.at[i, "numPasses"] = yanited_passingNetwork.loc[yanited_passingNetwork["player_name"] == name].count().iloc[0]

passingNetwork_df['marker_size'] = (passingNetwork_df['numPasses'] / passingNetwork_df['numPasses'].max() * 1500)
```

```python
passingNetwork_df
```

```python
lines_df = yanited_passingNetwork.groupby(
    ['player_name','pass_recipient_name']).x.count().sort_values(
    ascending=False).reset_index()
lines_df.rename({'x':'pass_count'},axis='columns',inplace=True)
lines_df = lines_df[lines_df['pass_count']>2]
```

```python
lines_df
```

```python
pitch = Pitch(pitch_type='statsbomb', line_zorder=2, 
              pitch_color='#45a15d', line_color='#efefef')
fig,ax = pitch.grid(figheight=10, title_height=0.08, endnote_space=0, 
                  axis=False,title_space=0, grid_height=0.82,
                  endnote_height=0.05)
pitch.scatter(passingNetwork_df.x,passingNetwork_df.y,
             s=passingNetwork_df.marker_size,c='red',alpha=1,
             ax=ax['pitch'])
for i,row in passingNetwork_df.iterrows():
    pitch.annotate(row.player_name,xy=(row.x,row.y),c='black',
                  ax=ax["pitch"],weight='bold')
for i,row in lines_df.iterrows():
    p1 = row['player_name']
    p2 = row['pass_recipient_name']
    p1_x = passingNetwork_df[passingNetwork_df['player_name']==p1].x.iloc[0]
    p1_y = passingNetwork_df[passingNetwork_df['player_name']==p1].y.iloc[0]
    p2_x = passingNetwork_df[passingNetwork_df['player_name']==p2].x.iloc[0]
    p2_y = passingNetwork_df[passingNetwork_df['player_name']==p2].y.iloc[0]
    numPasses = row['pass_count']
    line_width = (numPasses / lines_df['pass_count'].max() * 10)
    pitch.lines(p1_x, p1_y, p2_x, p2_y,
                        alpha=1, lw=line_width, zorder=2, color="red", ax = ax["pitch"])
```
