```python
import re
import json
import requests
import pandas as pd
import warnings
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,make_scorer
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn.model_selection import GridSearchCV, KFold
```

```python
warnings.filterwarnings('ignore')

seasons = [2014,2015,2016,2017,2018,2019,2020,2021,2022]
competitions = ['EPL','La_liga','Serie_A','Bundesliga','Ligue_1','RFPL']

all_data = []
for season in seasons:
    for comp in competitions:
        url = f"https://understat.com/league/{comp}/{season}"
        html_doc = requests.get(url).text

        data = re.search(r"datesData\s*=\s*JSON\.parse\('(.*?)'\)", html_doc).group(1)
        data = re.sub(r'\\x([\dA-F]{2})', lambda g: chr(int(g.group(1), 16)), data)
        data = json.loads(data)

        for d in data:
            all_data.append({
                'season': season,
                'competition': comp,
                'date': d['datetime'][:10], # first ten letters
                'home_team': d['h']['title'],
                'away_team': d['a']['title'],
                'home_goals': d["goals"]["h"],
                'away_goals': d["goals"]["a"],
                'home_xG':d['xG']['h'],
                'away_xG': d['xG']['a'],
                'forecast': list(d.get('forecast', {}).values())
            })

df = pd.DataFrame(all_data)
# Split the forecast list into separate columns
df[['home_win_prob', 'draw_prob', 'away_win_prob']] = df['forecast'].apply(lambda x: pd.Series(x))

# Drop the original forecast column
df = df.drop('forecast', axis=1)

# Drop the games that haven't been played
df = df.dropna(how='any', subset=None)

df.to_csv('xg_model.csv',index=False)
```

```python
class DataPreprocessing:
    
    def __init__(self,Dataframe):
        self.df = Dataframe
         #interchangably pass list of features ['home_team','away_team','home_xg','away_xg','home_goals','away_goals']
        self.df = self.df.iloc[:, [0] + list(range(3, 9))]
     
    
    def LabelData(self):
        # Creation of target/label column --> result
        result = []
        for ind,row in self.df.iterrows():
            if row.home_goals > row.away_goals:
                result.append('home')
            elif row.away_goals > row.home_goals:
                result.append('away')
            else:
                result.append('draw')
        
        self.df['result'] = result
        return self.df
    
    
    # dictionary -- Team Name as key, list of [brH, brA, form_count] as value
    def Team_BR_FormCount_SearchDict(self):
        # Check if search_dict already exists as an attribute
        if hasattr(self, 'search_dict'):
            print('already there\n')
            return self.search_dict
        else:
            # Create the search_dict and assign it as an attribute
            self.search_dict = dict.fromkeys(self.df.home_team.unique(), [0, 0, 0])
            return self.search_dict
    
    


class OEE:
        
    @staticmethod
    def GoalObservedDiff(gH, gA):
        gO = np.abs(gH - gA)
        return gO
    
    
    @staticmethod
    def GoalExpectedDiff(xGH,xGA):
        gP = np.abs(xGH - xGA)
        return gP
    
    
    @staticmethod
    def error(gO,gP):
        error = np.abs(gO - gP)
        return error
    



class Dolores(OEE):
    
    def __init__(self, lambda_lr1, c, gamma_lr2, form_threshold, provisional_rating_impact, form_diminishing_factor):
        self.lambda_lr1 = lambda_lr1
        self.gamma_lr2 = gamma_lr2
        self.c = c
        self.form_threshold = form_threshold
        self.provisional_rating_impact = provisional_rating_impact
        self.form_diminishing_factor = form_diminishing_factor
    
    pass




class Ratings(DataPreprocessing, Dolores, OEE):
    
    pass
    def __init__(self,Dataframe,lambda_lr1, c, gamma_lr2, form_threshold, 
                 provisional_rating_impact, form_diminishing_factor):
        # Create the necessary preprocessed dataframe from the DataPreprocessing class constructor
        super.__init__(Dataframe)
        # Create the required target labels from the Datapreprocessing class method 'LabelData()'
        self.LabelData()
        
        
        
    
    
```
